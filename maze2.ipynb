```python
import numpy as np
import gym
from IPython import display
import matplotlib.pyplot as plt
from utils import plot_policy, plot_values, test_agent
from envs import Maze
%matplotlib inline
```

```python
def random_policy(env):
    return np.array([0.25]*4)
env = Maze()
state = env.reset()
```

```python
action_probability = random_policy(env)
object = ('Up','Right','Down','Left')
y_pos = np.arange(len(object))
```

```python
plt.bar(y_pos, action_probability, alpha=0.5)
plt.xticks(y_pos, object)
plt.ylabel('P(a|s)')
plt.title('Random Policy')
plt.tight_layout()
plt.show()
```

```python
env.reset()
done = False
img = plt.imshow(env.render(mode='rgb_array'))
while not done:
    action = np.random.choice(range(4),1,p=action_probability)
    _,_,done,_ = env.step(action)
    img.set_data(env.render(mode='rgb_array'))
    plt.axis('off')
    display.display(plt.gcf())
    display.clear_output(wait=True)
env.close()
```
